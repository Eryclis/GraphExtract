```python
import pandas as pd
```

```python
import os
```

```python
path="out/"
files = os.listdir(path)
nodes=set()
edges=set()
for f in files:
    temp=pd.read_csv("out/"+f, sep="|")
    nodes.update(temp["node_1"])
    nodes.update(temp["node_2"])
    edges.update(temp["edge"])
```

```python
len(nodes)
```

```python
edges=list(edges)
nodes=list(nodes)
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim

#sentences = ['coauthor', 'collaborator']
model = SentenceTransformer('thenlper/gte-small')

```

```python
embeddings = model.encode(nodes)
#print(cos_sim(embeddings[0], embeddings[1]))
```

```python
res = dict(map(lambda i,j : (i,j) , nodes,embeddings))
```

```python
import pickle

filehandler = open("nodes_dict.obj","wb")
pickle.dump(res,filehandler)
filehandler.close()
```

```python
path="semantic/"
files = os.listdir(path)
text_emb=dict()
for f in files:
    file = open("semantic/"+f, "r", encoding="utf-8")
    content = file.read()
    file.close()
    id=f.split(".")[0]
    texts=[content[i:i+2500] for i in range(0, len(content), 2500)]
    embeddings = model.encode(texts)
    tot=[0]*384
    for e in embeddings:
        tot+=e
    text_emb[id]=tot
```

```python
df=pd.DataFrame.from_dict(text_emb)
```

```python
df.T.to_csv("doc_embeddings.csv")
```
